###  4. Cleaning data

**The CarName field has Company name and model name concatenated. we can analyse the componay name and the average price of the vehicles the company markets to see which brand segment influences the pricing of the vehicles**

```python
#Check all the unique values in the field CarName
CarPrice_data['CarName'].unique()
```

```python
#Get company names
CarPrice_data['CarName'].apply(lambda x : x.split(' ')[0]).unique()
```

```python
CarPrice_data['company'] = CarPrice_data['CarName'].apply(lambda x : x.split(' ')[0])
```

**We have some car manufacturer names misspelled or used abbrevations, so we will change those names to reflect correct names.**


```python
name_change = {"company": {"maxda": "mazda", "nissan": "Nissan", "toyouta": "toyota", "porcshce": "porsche",
                                  "vokswagen": "volkswagen", "vw": "volkswagen" }}
CarPrice_data.replace(name_change, inplace=True)
CarPrice_data.head()
```

```python
plt.figure(figsize=(20, 20))
plt.subplot(3,3,1)
plt.xticks(rotation=90);
sns.boxplot(x = 'company', y = 'price', data = CarPrice_data)
plt.show()
```

**We can create a categorical variable segment based on average price of the brand**

 - mean(price) < 10K = low-range
 - mean(price) >= 10 K and < 20K = lower-mid-range
 - mean(price) >= 20 K and < 30K = upper-mid-range
 - mean(price) >= 30 K  = luxury-range
 

```python
compony_mean_price = CarPrice_data.groupby('company')['price'].mean().reset_index()
```

```python
compony_mean_price['segment'] = compony_mean_price['price'].apply(lambda x : 'low-range' if x < 10000 else
                                 ("lower-mid-range" if (x >=10000 and x < 20000) else
                                  ("upper-mid-range" if (x >=20000 and x < 30000) else "luxury-range"
                                 
                                 
                                )))
```

```python
CarPrice_data = pd.merge(CarPrice_data,compony_mean_price, on = 'company', how = 'left')
# Drop 'fueltype' as we have created the dummies for it

CarPrice_data.drop(['company','CarName','price_y'], axis = 1, inplace = True)
CarPrice_data.head()
```

### fuelsystem
**For fuel sysetm mfi and spfi are subcategories of mpfi, so we can merge them to mpfi**

```python
CarPrice_data['fuelsystem'] = CarPrice_data['fuelsystem'].apply(lambda x:"mpfi" if x in ("mfi","spfi") else x )
```

### 5. Data preparation

**Convert numbers in words to numbers**

```python
#Convert numbers in words to numbers
NumList = ['doornumber', 'cylindernumber']

for i in NumList:
    CarPrice_data[i] = CarPrice_data[i].apply(lambda x: w2n.word_to_num(x) ) 

#Display head
CarPrice_data.head()
```

**Create categories for variable symboling**

```python
CarPrice_data['symboling_cat'] = CarPrice_data['symboling'].apply(lambda x: "Low_risk" if x < 0 
                                                                  else ("moderate_risk" if x in (0,1) else "high_risk") )
```

```python
#Drop symboling column as its not required
CarPrice_data.drop(['symboling'], axis = 1, inplace = True)
```

**As we have tow milage variables for citympg and highwaympg, we can create a derived column from these 2 columns**

```python
CarPrice_data['avg_mpg'] = (CarPrice_data['citympg'] + CarPrice_data['highwaympg'] )/2
```

```python
#Drop symboling column as its not required
CarPrice_data.drop(['citympg','highwaympg'], axis = 1, inplace = True)
```

### 6. Handling of Categorical Variables

**As we have multiple categorical variables, we will have to create dummy variables which can represent the categorical variables in the dataset**

```python
#Create a list of all categorical variables which we will convert into dummy variables

cat_list = ['fueltype','aspiration','carbody','drivewheel','enginelocation','enginetype','fuelsystem','segment','symboling_cat']
```

```python
#convert categorical variables into dummy variables
for var in cat_list:
        dummy_x = pd.get_dummies(CarPrice_data[var],drop_first= True)
        CarPrice_data = pd.concat([CarPrice_data, dummy_x], axis = 1)
        CarPrice_data.drop([var], axis = 1, inplace = True)
```

```python
#Drop car_id column as its not required
CarPrice_data.drop(['car_ID'], axis = 1, inplace = True)

#rename price_x to price
CarPrice_data.rename({'price_x':'price'},axis = 1, inplace = True)
```

### 10. Model Building

**I will use recursive feature elimination method to select top 10 featues**

```python
# Running RFE with the output number of the variable equal to 15
lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm, 15)             
rfe = rfe.fit(X_train, y_train)
```

```python
#display all variables and their ranking
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
#Display 10 selected features

col = X_train.columns[rfe.support_]
col
```

```python
#display features which we will not use for modelling
X_train.columns[~rfe.support_]
```

```python
# Creating X_test dataframe with RFE selected variables
X_train_rfe = X_train[col]
```

```python
#Create a function to apply RFE on the dataset

def apply_rfe(x_train_dataset, y_train_dataset):
    # Adding a constant variable as statsmodel does not add it by default
    x_train_dataset = sm.add_constant(x_train_dataset)
    
    # Running the linear model
    lm = sm.OLS(y_train_dataset,x_train_dataset).fit()  
    #Let's see the summary of our linear model
    print(lm.summary())
    
    #As hardtop has high p-value we need to drop this feature
    x_train_dataset = x_train_dataset.drop(['const'], axis=1)
    
    
```

```python
# create a function to calculate the VIFs for the model
def calc_vif(X_train_dataset):
    vif = pd.DataFrame()
    X = X_train_dataset
    vif['Features'] = X.columns
    vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    vif['VIF'] = round(vif['VIF'], 2)
    vif = vif.sort_values(by = "VIF", ascending = False)
    print(vif)
```
