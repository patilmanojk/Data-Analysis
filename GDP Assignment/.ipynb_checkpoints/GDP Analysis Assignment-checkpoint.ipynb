## Categorise the states into four categories based on GDP per capita (C1, C2, C3, C4 - C1 would have the highest per capita GDP, C4 the lowest)

```python
# Create Quantile Variable which can be used for categories creation

quantile_100 =  states[states['Item'] == 'Per Capita GSDP (Rs.)']['2014-15'].quantile(1)
quantile_85 =   states[states['Item'] == 'Per Capita GSDP (Rs.)']['2014-15'].quantile(0.85)
quantile_50 =   states[states['Item'] == 'Per Capita GSDP (Rs.)']['2014-15'].quantile(0.50)
quantile_20 =   states[states['Item'] == 'Per Capita GSDP (Rs.)']['2014-15'].quantile(0.20)
```

```python
# Function will return category, provided 'Per Capita GSDP (Rs.)' value
def category(row):
  if (row <= quantile_100) & (row> quantile_85):
    return 'C1'
  elif (row  <= quantile_85) & (row> quantile_50):
    return 'C2'
  elif (row <= quantile_50) & (row > quantile_20):
    return 'C3'
  else:
    return 'C4'
                                                                                 
```

```python
# create a new column 'category' 
states['category'] = states[states['Item'] == 'Per Capita GSDP (Rs.)']['2014-15'].apply(lambda x: category(x))
states_category = states[states['Item'] == 'Per Capita GSDP (Rs.)']
states = pd.merge(states[['index','S.No.','Item','2014-15','State','Sector']], 
                 states_category[['State','category']],how='inner', on=['State'])
```

## Data Cleaning

  - Select relevent data needed for 2014-2015 year analysis
  - clean the data so that corresponding data can be fetched from GDP dataset
  e.g.
      1.   replace spaces in the state names by '_'
      2.   Change 'Chhatisgarh' to 'Chhattisgarh'
      3.  Change 'Uttrakhand' to 'Uttarakhand'
 note: as column "Primary - 2014-2015" is duplicated, I have selected first entry in DataFrame which is "Primary - 2014-2015"
  

```python
#Select relevent data needed for 2014-2015 year analysis
drop_out_data_2014_15 = drop_out_data[['Level of Education - State', 'Primary - 2014-2015',  'Upper Primary - 2014-2015','Secondary - 2014-2015']].copy()
```

```python
# replace spaces in the state names by '_'
drop_out_data_2014_15['Level of Education - State'] = drop_out_data_2014_15['Level of Education - State'].apply(lambda x: x.replace(' ','_'))
```

```python
#Change 'Chhatisgarh' to 'Chhattisgarh'
drop_out_data_2014_15['Level of Education - State'] = drop_out_data_2014_15['Level of Education - State'].apply(lambda x: x.replace('Chhatisgarh','Chhattisgarh'))
```

```python
#Change 'Uttrakhand' to 'Uttarakhand'
drop_out_data_2014_15['Level of Education - State'] = drop_out_data_2014_15['Level of Education - State'].apply(lambda x: x.replace('Uttrakhand','Uttarakhand'))
```

```python
#select relevent data from GDP dataset "states" and create a new dataset "states_Per_Capita_GSDP" 
#which can be used for further analysis 
states_Per_Capita_GSDP  = states[states['Item'] =='Per Capita GSDP (Rs.)'][['State','2014-15','category']]
states_Per_Capita_GSDP.reset_index(inplace = True)
states_Per_Capita_GSDP.drop(columns = 'index', inplace = True )
```

```python
states_Per_Capita_GSDP.info()
```

## Join drop_out_data_2014_15 dataset with states_Per_Capita_GSDP to fetch relevent GDP per capita information

```python

GSDP_Dropout = pd.merge(states_Per_Capita_GSDP, drop_out_data_2014_15, left_on='State', right_on='Level of Education - State')
```

```python
# drop 'Level of Education - State' as it is duplicate information
GSDP_Dropout.drop(columns = ['Level of Education - State'], inplace= True)
```

```python
#rename "2014-15" to "GSDP_per_capita"
GSDP_Dropout.rename(columns = {"2014-15": "GSDP_per_capita"}, inplace= True)
```

```python
# create index with the 'State' Column
GSDP_Dropout.set_index('State',inplace=True)
```

```python
GSDP_Dropout.describe()
```
